## Exercise 2
- Let's turn this into a time series.

 For each key (basically the control area, unit, device address and station of a specific turnstile), have a list again, but let the list be comprised of just the point in time and the cumulative count of entries.

This basically means keeping only the date, time, and entries fields in each list. You can convert the date and time into datetime objects -- That is a python class that represents a point in time. You can combine the date and time fields into a string and use the [dateutil](https://dateutil.readthedocs.io/en/stable/) module to convert it into a datetime object.

Your new dict should look something like
 
    {    ('A002','R051','02-00-00','LEXINGTON AVE'):    
             [
                [datetime.datetime(2013, 3, 2, 3, 0), 3788],
                [datetime.datetime(2013, 3, 2, 7, 0), 2585],
                [datetime.datetime(2013, 3, 2, 12, 0), 10653],
                [datetime.datetime(2013, 3, 2, 17, 0), 11016],
                [datetime.datetime(2013, 3, 2, 23, 0), 10666],
                [datetime.datetime(2013, 3, 3, 3, 0), 10814],
                [datetime.datetime(2013, 3, 3, 7, 0), 10229],
                ...
              ],
     ....
     }



```python
import itertools
import datetime as datetime
import csv
import requests
req=requests.get('http://web.mta.info/developers/data/nyct/turnstile/turnstile_170107.txt')
data=req.text
data_CSV=csv.reader(data.splitlines(),delimiter=',')
next(data_CSV,None) # remove header
dict={}
for row in itertools.islice(data_CSV,100000):
    if tuple(row[0:4]) not in dict:
        row_slice=row[6:11]
        row_slice.pop(2)
        date=row_slice[0].split('/')
        date = [ int(x) for x in date ]
        time=row_slice[1].split(':')
        time = [ int(x) for x in time ]
        net_entries=int(row_slice[2])-int(row_slice[3])
        row_slice_mod=[datetime.datetime(date[2],date[0],date[1],time[0],time[1]),net_entries]
        dict[tuple(row[0:4])]=[row_slice_mod]
    else:
        row_slice=row[6:11]
        row_slice.pop(2)
        date=row_slice[0].split('/')
        date = [ int(x) for x in date ]
        time=row_slice[1].split(':')
        time = [ int(x) for x in time ]
        net_entries=int(row_slice[2])-int(row_slice[3])
        row_slice_mod=[datetime.datetime(date[2],date[0],date[1],time[0],time[1]),net_entries]
        dict[tuple(row[0:4])]=dict[tuple(row[0:4])]+[row_slice_mod]
#for keys,values in dict.items():
    #print(keys)
    #print(values)
```

## Exercise 3
- These counts are cumulative every n hours. We want total daily entries. 

Now make it that we again have the same keys, but now we have a single value for a single day, which is not cumulative counts but the total number of passengers that entered through this turnstile on this day.


```python
dict1=dict
dict2={}
for key in dict1:
    #print(dict1[key])
    subdict={}
    for item in dict1[key]:
        #print(item)
        date = item[0].date()
        if date not in subdict:
            subdict[date]=[item[1]]
        else:
            subdict[date]=subdict[date]+[item[1]]
    #dict1[key]=[[]]
    #print(subdict)
    for item1 in subdict:
        a=subdict[item1]
        if key not in dict2:
            dict2[key]=[[item1,max(a)-min(a)]]
        else:
            dict2[key]=dict2[key]+[[item1,max(a)-min(a)]]
#print(dict2)        
```

## Exercise 5
- So far we've been operating on a single turnstile level, let's combine turnstiles in the same ControlArea/Unit/Station combo. There are some ControlArea/Unit/Station groups that have a single turnstile, but most have multiple turnstilea-- same value for the C/A, UNIT and STATION columns, different values for the SCP column.

We want to combine the numbers together -- for each ControlArea/UNIT/STATION combo, for each day, add the counts from each turnstile belonging to that combo.


```python
dict3={}
dict4={}
for key in dict2:
    key2=(key[0],key[1],key[3])
    if key2 not in dict3:
        dict3[key2]=dict2[key]
    else:
        dict3[key2]=dict3[key2]+dict2[key]
#print(dict3)
#print('-------')
for key in dict3:
    subdict={}
    for item in dict3[key]:
        if item[0]not in subdict:
            subdict[item[0]]=item[1]
        else:
            subdict[item[0]]=subdict[item[0]]+item[1]
    for item in subdict:
        a=subdict[item]
        if key not in dict4:
            dict4[key]=[[item,a]]
        else:
            dict4[key]=dict4[key]+[[item,a]]
#print(dict4)        
```
